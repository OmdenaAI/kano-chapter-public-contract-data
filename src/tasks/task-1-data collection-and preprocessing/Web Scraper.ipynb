```python
###Loading required libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd 
import numpy as np
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time
import calendar
from datetime import date
from datetime import time
from datetime import datetime
#import schedule


from os.path import exists

import urllib.request
 
# pretty-print python data structures
from pprint import pprint
 
from html_table_parser.parser import HTMLTableParser
```

```python
#state = "lagos"
#state = "kaduna"
state = "FG"

#link = "https://lagosppaocds.azurewebsites.net/Home/Procurements"
link = "https://nocopo.bpp.gov.ng/OpenData.aspx"
```

```python
#function to get website contents
def url_get_contents(url):
 
    
    #to make request from the website
    req = urllib.request.Request(url=url)
    f = urllib.request.urlopen(req)
 
    #to read contents of the website
    return f.read()
```

```python
xhtml = url_get_contents(link).decode('utf-8')
xhtml
```

```python
#Load and feed scraped data into table parser
p = HTMLTableParser()
p.feed(xhtml)
```

```python
#load table
table = p.tables[0]

table
```

```python


#Load table into pandas dataframe
df = pd.DataFrame(table)
```

```python
df.columns = df.iloc[0]   #change name of name to first row in the columns
df
```

```python
df = df.iloc[1:, :]  #remove first row
df
```
