```python
import pandas as pd
```

```python
name = "Project Name"
price = "Price"
loc =  "Location"
ad = "Award Date"
cdate = "Completion Date"
dur = "Duration"
cl = "Client"
spon = "Sponsor"

def add_state(l, df, state):
    st = []
    for i in range(l):
        st.append(state)

    df["State"] = st
    return df
```

```python
df1 = pd.read_csv("Kaduna MFA.csv")
df1
```

```python
l = len(df1["PROJECT"])
df1 = add_state(l, df1, "Kaduna")
df1
```

```python
df1.rename(columns = {"PROJECT":name, "PROJECT COST":price, "CONTRACTOR":cl}, inplace = True)
df1
```

```python
df2 = pd.read_csv("lagos_ppa_data.csv")
df2
```

```python
df2 = df2.drop(["Unnamed: 0"], axis=1)
df2
```

```python
l = len(df2["Project Name"])
df2 = add_state(l, df2, "Lagos")
df2
```

```python
df2.rename(columns = {"Project Cost":price, "Contractor":cl, "Project Location":loc}, inplace = True)
df2
```
